# Stacey's Code for data clean up


```python
#import dependencies

import pandas as pd
import os
import csv
```

```python
# The path to our CSV file
file = "Resources/or_historical_wildfires.csv"

#Read data into csv
or_hist_df = pd.read_csv(file)
or_hist_df.head()
```

```python
#drop NAN values, we want to be able to plot with lat and lon
or_hist_df = or_hist_df.dropna(how='any')
```

```python
or_hist_df.head()
```

```python
#drop fuel description that is less than one acre and isn't a wildland fuel
index_names = or_hist_df[ (or_hist_df['total_acres'] <= .05) & (or_hist_df['fuel_descr'] == 'non-wildland fuel')].index
or_hist_df.drop(index_names, inplace = True)
```

```python
or_hist_df.head()
```

```python
#grab columns we care about
or_df = or_hist_df[['fire_year', 'report_date', 'county', 'latitude', 'longitude', 'total_acres', 'general_cause']]
```

```python
#sort by report date
or_df = or_df.sort_values(by=['report_date'])
```

```python
#reset index
or_df = or_df.reset_index(drop=True)
or_df
```

```python
or_df.to_csv("Resources/or_df.csv")
```
